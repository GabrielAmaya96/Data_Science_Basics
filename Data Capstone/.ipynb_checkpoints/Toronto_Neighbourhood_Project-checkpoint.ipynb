### 1. Get Wikipedia table and make it data mining

Through the use of BeautifulSoup library we are going to create a simple web scraping for Wikipedia table, after we are going to create a data frame and we are going to clean it with the objective to use this information with foursquare API.

```python
# Create a connection from wikipedia
Link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
Wiki_link = requests.get(Link)
soup = BeautifulSoup(Wiki_link.content, 'html.parser')

# Scrape the table we need
table = soup.find_all("table", attrs={'class': 'wikitable'})
print(table)
```

```python
# Extract information about table that begins and end with tr
table = table[0]
content = table.find_all('tr')

# Get the information from header and body.
head = content[0] # This give us the headers of the table.
body_rows = content[1:] # This give us the content of the table.

# Create a loop to clean headers from HTML format

# Declare empty list to keep Columns names
headings = []
for item in head.find_all("th"): # loop through all th elements
    # convert the th elements to text and strip "\n"
    item = (item.text).rstrip("\n")
    # append the clean column name to headings
    headings.append(item)
print(headings)
```

```python
# Create a loop to clean and get the body information

all_rows = [] # will be a list for list for all rows
for row_num in range(len(body_rows)):
    row = []
    for row_item in body_rows[row_num].find_all("td"):
        # row_item.text removes the tags from the entries
        # the following regex is to remove \xa0 and \n from row_item.text
        # xa0 encodes the flag, \n is the newline and comma separates thousands in numbers
        aa = re.sub("(\xa0)|(\n)|","",row_item.text)
        #append aa to row - note one row entry is being appended
        row.append(aa)
    # append one row to all_rows
    all_rows.append(row)
```

```python
# all_rows becomes our data and headings the column names to create a dataframe
Toronto = pd.DataFrame(data=all_rows,columns=headings)
Toronto.head()
```

### 3. Create a cluster analysis by similar postal codes.
Use the Foursquare API we get the information about places near to postal codes and create clusters based on it information, we are going to work only with Boroughs that has the word 'Toronto'

```python
# Filter the boroughs with toronto name
Borough_Toronto = Toronto[Toronto['Borough'].str.contains('Toronto')].reset_index(drop = True)
print(Borough_Toronto.shape)
Borough_Toronto.head()
```

```python
# Define credentials for all the process to work with 
CLIENT_ID = 'Client ID' # your Foursquare ID
CLIENT_SECRET = 'Client Secret' # your Foursquare Secret
VERSION = 'Version API' # Foursquare API version
LIMIT = 100 # A default Foursquare API limit value

print('Your are ready to work!')
```

```python
# Define a function to get information about near places of each postal codes
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Postal Code', 
                  'Postal Code Latitude', 
                  'Postal Code Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# Get the information about each borough based on latitude and longitude
LIMIT = 100
Toronto_venues = getNearbyVenues(names=Borough_Toronto['Postal Code'],
                                latitudes=Borough_Toronto['Latitude'],
                                longitudes=Borough_Toronto['Longitude'],
                                radius = 500
                                )
```

```python
# Lets check the information we get about each postal code.
print(Toronto_venues.shape)
Toronto_venues.head()
```

```python
# Understand information we get, first we merge the borough and Neighbourhood
Neig_Toronto = Borough_Toronto[['Postal Code','Borough','Neighbourhood']]
Toronto_data = pd.merge(left = Toronto_venues, right = Neig_Toronto, on = 'Postal Code')
Toronto_data.head()
```

```python
# Create a groupby to look the quantitive of information we get by Neighbourhood
Toronto_data.groupby('Neighbourhood').count()
```

```python
# Create a one hot encoding data frame to look main characteristics
Toronto_onehot = pd.get_dummies(Toronto_data[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Toronto_onehot['Neighbourhood'] = Toronto_data['Neighbourhood'] 

# move neighborhood column to the first column
fixed_columns = [Toronto_onehot.columns[-1]] + list(Toronto_onehot.columns[:-1])
Toronto_onehot = Toronto_onehot[fixed_columns]

# Review the new data frame transformation
print(Toronto_onehot.shape)
Toronto_onehot.head()
```

```python
# Create a groupby by each Neighbourhood based on mean of each characteristics
Toronto_grouped = Toronto_onehot.groupby('Neighbourhood').mean().reset_index()
Toronto_grouped.head()
```

```python
# Review the top 5 common venues on each characteristics.
num_top_venues = 5

# Create a loop to see main characteristics.
for hood in Toronto_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = Toronto_grouped[Toronto_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# Create a function with the most common venues
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# Define the top of venues we want to see in the data frame.
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
Toronto_sorted = pd.DataFrame(columns=columns)
Toronto_sorted['Neighbourhood'] = Toronto_grouped['Neighbourhood']

for ind in np.arange(Toronto_grouped.shape[0]):
    Toronto_sorted.iloc[ind, 1:] = return_most_common_venues(Toronto_grouped.iloc[ind, :], num_top_venues)

Toronto_sorted.head()
```

```python
# set number of clusters
kclusters = 10

Toronto_grouped_clustering = Toronto_grouped.drop('Neighbourhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# add clustering labels
Toronto_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

Toronto_merged = Toronto_data

# merge toronto grouped with toronto_data to add latitude/longitude for each neighborhood
Toronto_merged = Toronto_merged.join(Toronto_sorted.set_index('Neighbourhood'), on='Neighbourhood')

Toronto_merged.head()
```

```python
# create map of toronto
latitude = 43.7166
longitude = -79.3407
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Toronto_merged['Postal Code Latitude'], Toronto_merged['Postal Code Longitude'], Toronto_merged['Neighbourhood'], Toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
